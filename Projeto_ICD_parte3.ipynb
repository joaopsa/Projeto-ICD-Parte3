```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(20,7)})

dados = pd.read_csv("twitchdata-update.csv")
dados
```

## Visualização de Dados

### Gráfico de Dispersão:

#### Esse gráfico mostra a dispersão de Seguidores com Seguidores ganhos(mês).

```python
sns.scatterplot(data=dados, x="Followers", y="Followers gained")
```

### Gráfico de Histograma:

#### Mostra a quantidade streamers que usam determinadas línguas.

```python
sns.displot(data=dados, x="Language", height=18, aspect=.9, hue="Language",)
```

### Gráfico de Barra:

#### Esse gráfico mostra a média de espectadores comparado de acordo com o idioma falado pelo Streamer.

```python
plt.figure(figsize=(20,7))
ax = sns.barplot(data=dados, x="Language", y="Average viewers")
ax.set(xlabel='Language', ylabel='Average viewers')
```
